```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# load the data with the frequency up to 200Hz
AMRRauschspektrum200InTonne = np.loadtxt('data/Leistungsspektrum_AMR_bis200Hz_inTonne.lvm')
AMRRauschspektrum200OhneTonne = np.loadtxt('data/Leistungsspektrum_AMR_bis200Hz_ausserhalbTonne.lvm')
```

```python
plt.figure()
plt.grid()
plt.semilogy(AMRRauschspektrum200OhneTonne[:,0], AMRRauschspektrum200OhneTonne[:,1], color='r', alpha=0.6, label = "Sensor au√üerhalb der Abschirmtonne")
plt.semilogy(AMRRauschspektrum200InTonne[:,0], AMRRauschspektrum200InTonne[:,1], alpha=0.6, label = "Sensor in der Abschirmtonne")

plt.legend(loc='upper right')
plt.xlabel('Frequenz in Hz')
plt.ylabel('Spannungsleistungsspektrum in $\mathrm{V^{2}/Hz}$')
plt.savefig('output/AMR_Rauschspektrum_200Hz.eps', format='eps', dpi=100)
plt.show()
```

```python
# load the data with the frequency up to 20kHz
AMRRauschspektrum20kInTonne = np.loadtxt('data/Leistungsspektrum_AMR_bis20000Hz_inTonne.lvm')
AMRRauschspektrum20kOhneTonne = np.loadtxt('data/Leistungsspektrum_AMR_bis20000Hz_ausserhalbTonne.lvm')
```

```python
plt.figure()
plt.grid()
plt.semilogy(AMRRauschspektrum20kOhneTonne[:,0], AMRRauschspektrum20kOhneTonne[:,1], color='r', alpha=0.6, label="Sensor au√üerhalb der Abschirmtonne")
plt.semilogy(AMRRauschspektrum20kInTonne[:,0], AMRRauschspektrum20kInTonne[:,1], alpha=0.6, label="Sensor in der Abschirmtonne")

plt.xlabel('Frequenz in Hz')
plt.ylabel('Spannungsleistungsspektrum in $\mathrm{V^{2}/Hz}$')
plt.legend(loc="upper right")
plt.savefig('output/AMR_Rauschspektrum_20kHz.eps', format='eps', dpi=100)
plt.show()
```

```python
plt.figure()
plt.grid()

plt.semilogy(AMRRauschspektrum20kOhneTonne[:,0], AMRRauschspektrum20kOhneTonne[:,1], color='r', alpha=0.6)

plt.xlabel('Frequenz in $\it{Hz}$')
plt.ylabel('Spannungsleistungsspektrum in $\it{V^{2}/Hz}$')
plt.savefig('output/AMR_Rauschspektrum_20kHz_ausserhalb_Tonne.eps', format='eps', dpi=100)
plt.show()
```

```python
# load the data of the spektrum elektronic
ElektronikRauschspektrum200 = np.loadtxt('data/Leistungsspektrum_Elektronik_bis200Hz.lvm')
ElektronikRauschspektrum20000 = np.loadtxt('data/Leistungsspektrum_Elektronik_bis20000Hz.lvm')
```

```python
plt.figure()
plt.grid()

plt.semilogy(AMRRauschspektrum200InTonne[:,0], AMRRauschspektrum200InTonne[:,1], color='b', alpha=0.6, label="Gesamtrauschen innerhalb der Tonne")
plt.semilogy(ElektronikRauschspektrum200[:,0], ElektronikRauschspektrum200[:,1], color='r', alpha=0.6, label="Ausleseelektronik")

plt.xlabel('Frequenz in Hz')
plt.ylabel('Spannungsleistungsspektrum in $\mathrm{V^{2}/Hz}$')
plt.legend(loc="upper right")
plt.savefig('output/Elektronik_Rauschspektrum_200Hz_in.eps', format='eps', dpi=100)
plt.show()
```

```python
plt.figure()
plt.grid()

plt.semilogy(AMRRauschspektrum20kInTonne[:,0], AMRRauschspektrum20kInTonne[:,1], color='b', alpha=0.6, label="Gesamtrauschen innerhalb der Tonne")
plt.semilogy(ElektronikRauschspektrum20000[:,0], ElektronikRauschspektrum20000[:,1], color='r', alpha=0.6, label="Ausleseelektronik")
plt.xlabel('Frequenz in Hz')
plt.ylabel('Spannungsleistungsspektrum in $\mathrm{V^{2}/Hz}$')
plt.legend(loc="upper right")
plt.savefig('output/Elektronik_Rauschspektrum_20000Hz_in.eps', format='eps', dpi=100)
plt.show()
```
