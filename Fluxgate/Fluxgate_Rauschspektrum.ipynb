```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# load the data with the frequency up to 200Hz
fluxgateRauschspektrum200InTonne = np.loadtxt('data/Leistungsspektrum_Fluxgate_bis200Hz_inTonne.lvm')
fluxgateRauschspektrum200OhneTonne = np.loadtxt('data/Leistungsspektrum_Fluxgate_bis200Hz_ausserhalbTonne.lvm')
```

```python
plt.figure()
plt.grid()
plt.semilogy(fluxgateRauschspektrum200OhneTonne[:,0], fluxgateRauschspektrum200OhneTonne[:,1], color='r', alpha=0.6, label = "Sensor au√üerhalb der Abschirmtonne")
plt.semilogy(fluxgateRauschspektrum200InTonne[:,0], fluxgateRauschspektrum200InTonne[:,1], alpha=0.6, label = "Sensor in der Abschirmtonne")

plt.xlabel('Frequenz in Hz')
plt.ylabel('Spannungsleistungsspektrum in $\mathrm{V^{2}/Hz}$')
plt.legend(loc='upper right')
plt.savefig('output/Fluxgate_Rauschspektrum_200Hz.eps', format='eps', dpi=100)
plt.show()
```

```python
# load the data with the frequency up to 20kHz
fluxgateRauschspektrum20kInTonne = np.loadtxt('data/Leistungsspektrum_Fluxgate_bis20000Hz_inTonne.lvm')
fluxgateRauschspektrum20kOhneTonne = np.loadtxt('data/Leistungsspektrum_Fluxgate_bis20000Hz_ausserhalbTonne.lvm')
```

```python
plt.figure()
plt.grid()
plt.semilogy(fluxgateRauschspektrum20kOhneTonne[:,0], fluxgateRauschspektrum20kOhneTonne[:,1], color='r', alpha=0.6, label = "Sensor au√üerhalb der Abschirmtonne")
plt.semilogy(fluxgateRauschspektrum20kInTonne[:,0], fluxgateRauschspektrum20kInTonne[:,1], alpha=0.6, label = "Sensor in der Abschirmtonne")


plt.xlabel('Frequenz in Hz')
plt.ylabel('Spannungsleistungsspektrum in $\mathrm{V^{2}/Hz}$')
plt.legend(loc='upper right')
plt.savefig('output/Fluxgate_Rauschspektrum_20kHz.eps', format='eps', dpi=100)
plt.show()
```

```python
plt.figure()
plt.grid()

plt.semilogy(fluxgateRauschspektrum20kOhneTonne[:,0], fluxgateRauschspektrum20kOhneTonne[:,1], color='r', alpha=0.6)
plt.ylim(10E-15, 10E-5)
plt.xlabel('Frequenz in Hz')
plt.ylabel('Spannungsleistungsspektrum in $\mathrm{V^{2}/Hz}$')
plt.savefig('output/Fluxgate_Rauschspektrum_20kHz_ausserhalb_Tonne.eps', format='eps', dpi=100)
plt.show()
```

```python
abschirmFaktor_200 = fluxgateRauschspektrum200InTonne[:,1] / fluxgateRauschspektrum200OhneTonne[:,1]
plt.figure()
plt.grid()
plt.semilogy(fluxgateRauschspektrum200OhneTonne[:,0], abschirmFaktor_200, alpha=0.6)
plt.xlabel('Frequenz in $\it{Hz}$')
plt.ylabel('Abschirmfaktor')
plt.savefig('output/Fluxgate_Abschirmfaktor_200.eps', format='eps', dpi=100)
plt.show()
```

```python
abschirmFaktor_20k = fluxgateRauschspektrum20kInTonne[:,1] / fluxgateRauschspektrum20kOhneTonne[:,1]
plt.figure()
plt.grid()
plt.semilogy(fluxgateRauschspektrum20kOhneTonne[:,0], abschirmFaktor_20k, alpha=0.6)
plt.xlabel('Frequenz in $\it{Hz}$')
plt.ylabel('Abschirmfaktor')
plt.savefig('output/Fluxgate_Abschirmfaktor_20k.eps', format='eps', dpi=100)
plt.show()
```
