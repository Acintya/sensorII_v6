```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# load the data with the frequency up to 200Hz
fluxgateRauschspektrum200InTonne = np.loadtxt('data/Leistungsspektrum_Fluxgate_bis200Hz_inTonne.lvm')
fluxgateRauschspektrum200OhneTonne = np.loadtxt('data/Leistungsspektrum_Fluxgate_bis200Hz_ausserhalbTonne.lvm')
```

```python
plt.figure()
plt.grid()
plt.semilogy(fluxgateRauschspektrum200InTonne[:,0], fluxgateRauschspektrum200InTonne[:,1], alpha=0.6)
plt.semilogy(fluxgateRauschspektrum200OhneTonne[:,0], fluxgateRauschspektrum200OhneTonne[:,1], color='r', alpha=0.6)

plt.xlabel('Frequenz in $\it{Hz}$')
plt.ylabel('Spannungsleistungsspektrum in $\it{V^{2}/Hz}$')
plt.savefig('output/Fluxgate/Fluxgate_Rauschspektrum_200Hz.eps', format='eps', dpi=100)
plt.show()
```
